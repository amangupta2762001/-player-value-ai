```python
import pandas as pd

data = {
    'age': [22, 28, 19, 30, 25, 27, 24, 21, 29, 23],
    'goals': [10, 15, 8, 5, 11, 7, 13, 6, 9, 12],
    'assists': [5, 6, 3, 2, 4, 5, 7, 1, 3, 6],
    'appearances': [25, 30, 20, 28, 27, 26, 30, 19, 29, 24],
    'club_rank': [1, 2, 5, 3, 2, 4, 1, 6, 3, 2],
    'position': ['Forward', 'Midfielder', 'Forward', 'Defender', 'Forward',
                 'Midfielder', 'Forward', 'Defender', 'Midfielder', 'Forward'],
    'market_value': [60, 50, 40, 30, 55, 45, 65, 25, 48, 58]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

# Convert position (text) to numbers using label encoding
le = LabelEncoder()
df['position_encoded'] = le.fit_transform(df['position'])
```

```python
df
```

```python
# Define features (X) and target/label (y)
X = df[['age', 'goals', 'assists', 'appearances', 'club_rank', 'position_encoded']]
y = df['market_value']

# Train the model using all data (no test split for now)
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

print("Model trained on full dataset.")

```

```python
# Predict market value for a new player
# Example: 24 years, 12 goals, 5 assists, 28 appearances, top club, Forward

sample = [[24, 12, 5, 28, 1, le.transform(['Forward'])[0]]]

prediction = model.predict(sample)
print("Predicted Market Value (€ million):", round(prediction[0], 2))
```

```python
import pandas as pd

# Create the input as a DataFrame (to match training format)
sample_df = pd.DataFrame([{
    'age': 24,
    'goals': 12,
    'assists': 5,
    'appearances': 28,
    'club_rank': 1,
    'position_encoded': le.transform(['Forward'])[0]
}])

# Make prediction
prediction = model.predict(sample_df)
print("Predicted Market Value (€ million):", round(prediction[0], 2))
```

```python
import joblib

joblib.dump(model, 'player_value_model.pkl')
joblib.dump(le, 'label_encoder.pkl')

```
